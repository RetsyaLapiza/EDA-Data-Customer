```python
data = pd.read_csv("contacts-4fab6196-5801-4a39-8be2-16459b8e1406.csv")
data.head(5).transpose()
```

```python
data = data.rename(columns={'Website':'Business Sector', 'Business  Sector':'Syarat Pembayaran',
                     'Syarat Pembayaran':'Produk Diminati','Produk Diminati':'Job Tittle',
                     'Job Tittle':'Status'})
```

```python
data.shape
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data.sample(7).transpose()
```

```python
data.dtypes
```

```python
data['Date Created'] = data['Date Created'].astype(np.datetime64)
data['Date Modified'] = data['Date Modified'].astype(np.datetime64)
```

```python
d1 = data.drop(data.iloc[:, [1,2,3,4,5,7]],axis = 1)
d1.sample(7).transpose()
```

```python
for a in d1.columns:
  print(d1[a].unique())
  print(d1[a].value_counts())
```

```python
d1['City'] = d1['City'].str.lower()
d1['Business Sector'] = d1['Business Sector'].str.lower()
```

```python
d1['Business Sector'].replace(['sub kontractor','sub-kontractor'],'sub kontraktor', inplace = True)
d1['Business Sector'].replace(['pemakai kaca','toko kaca ','toko kaca + pemasangan','toko kaca'], 'toko kaca', inplace = True)
d1['Business Sector'].replace(['toko rak piring', 'toko furniture/meuble','toko besi'], 'lainnya', inplace = True)
d1['Business Sector'].replace(['arsitek', 'eksportir'], 'kontraktor', inplace =True)
```

```python
d1['City'].replace(['jakbar', 'west jakarta', 'kalideres'],'jakarta barat', inplace =True)
d1['City'].replace(['jakut', 'north jakarta'],'jakarta utara', inplace =True)
d1['City'].replace(['jakpus', 'central jakarta'],'jakarta pusat', inplace =True)
d1['City'].replace(['jaktim', 'east jakarta'],'jakarta timur', inplace =True)
d1['City'].replace(['jaksel', 'south jakarta'],'jakarta selatan', inplace =True)
d1['City'].replace(['depok', 'kota depok'],'depok', inplace =True)
d1['City'].replace(['south tangerang', 'ciputat - tangerang'],'tangerang selatan', inplace =True)
```

```python
print(d1['Produk Diminati'].unique())
print(d1['Produk Diminati'].value_counts())
```

```python
d1.loc[269, 'Produk Diminati']
```

```python
d1['Produk Diminati'] = d1['Produk Diminati'].str.replace("\W", "")
```

```python
d1['Produk Diminati'] = d1['Produk Diminati'].str.replace("\[IGU]","Igu")
```

```python
d1['Produk Diminati'] = d1['Produk Diminati'].str.replace(r"([A-Z])", r" \1", regex = True)
```

```python
print(d1['Produk Diminati'].value_counts())
```

```python
d1['Produk Diminati'].sample(25)
```

```python
d1[['Produk 0','Produk 1','Produk 2','Produk 3','Produk 4',
    'Produk 5','Produk 6','Produk 7','Produk 8',
    'Produk 9','Produk 10','Produk 11']] = d1['Produk Diminati'].str.split(' ', n = 11, expand = True)
```

```python
d1.drop(['Produk 0', 'Produk Diminati'], axis = 1, inplace = True)
```

```python
d3 = d2.dropna(subset = ['City', 'Business Sector'])
print(d3.isnull().sum())
print(d3.shape)
```

```python
d3 = d1.drop(['Date Created', 'Date Modified', 'Produk 3', 'Produk 4', 'Produk 5', 'Produk 6', 'Produk 7', 'Produk 8', 'Produk 9', 'Produk 10', 'Produk 11'], axis =1)
```

```python
d3['City'].fillna('lainnya',inplace = True)
d3['Salutation'].fillna('Tidak diketahui', inplace = True)
d3['Business Sector'].fillna('Tidak diketahui', inplace = True)
d3['Syarat Pembayaran'].fillna('Tidak diketahui', inplace = True)
d3['Job Tittle'].fillna('Tidak diketahui', inplace = True)
d3['Status'].fillna('Existing Customer', inplace = True)
d3['Produk 1'].fillna('Tidak diketahui', inplace = True)
d3['Produk 2'].fillna('Tidak diketahui', inplace = True)
d3.sample(7).transpose()
```
